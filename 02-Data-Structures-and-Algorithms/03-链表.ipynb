```python
class SingleLinkList(object):
    """"""
    def __init__(self):
        self._head = None

    def is_empty(self):
        """"""
        return self._head == None

    def length(self):
        """"""
        # cur
        cur = self._head
        count = 0
        # None
        while cur != None:
            count += 1
            # cur
            cur = cur.next
        return count

    def travel(self):
        """"""
        cur = self._head
        while cur != None:
            print(cur.item)
            cur = cur.next
        print("")

    def add(self, item):
        """"""
        # item
        node = SingleNode(item)
        # next_head
        node.next = self._head
        # _head
        self._head = node

    def append(self, item):
        """"""
        node = SingleNode(item)
        # _head
        if self.is_empty():
            self._head = node
        # next
        else:
            cur = self._head
            while cur.next != None:
                cur = cur.next
            cur.next = node

    def insert(self, pos, item):
        """"""
        # pos
        if pos <= 0:
            self.add(item)
        # 
        elif pos > (self.length() - 1):
            self.append(item)
        # 
        else:
            node = SingleNode(item)
            count = 0
            # prepospos-1
            pre = self._head
            while count < (pos - 1):
                count += 1
                pre = pre.next
            # nodenext
            node.next = pre.next
            # next
            pre.next = node

    def remove(self, item):
        """"""
        cur = self._head
        pre = None
        while cur != None:
            # 
            if cur.item == item:
                # 
                if not pre:
                    # 
                    self._head = cur.next
                else:
                    # next
                    pre.next = cur.next
                break
            else:
                # 
                pre = cur
                cur = cur.next

    def search(self, item):
        """TrueFalse"""
        cur = self._head
        while cur != None:
            if cur.item == item:
                return True
            cur = cur.next
        return False
```

```python
class Node(object):
    """"""
    def __init__(self, item):
        self.item = item
        self.next = None
        self.prev = None


class DLinkList(object):
    """"""
    def __init__(self):
        self._head = None

    def is_empty(self):
        """"""
        return self._head == None

    def length(self):
        """"""
        cur = self._head
        count = 0
        while cur != None:
            count += 1
            cur = cur.next
        return count

    def travel(self):
        """"""
        cur = self._head
        while cur != None:
            print(cur.item, end=" ")
            cur = cur.next
        print("")

    def add(self, item):
        """"""
        node = Node(item)
        if self.is_empty():
            # _headnode
            self._head = node
        else:
            # nodenext_head
            node.next = self._head
            # _headprevnode
            self._head.prev = node
            # _head node
            self._head = node

    def append(self, item):
        """"""
        node = Node(item)
        if self.is_empty():
            # _headnode
            self._head = node
        else:
            # 
            cur = self._head
            while cur.next != None:
                cur = cur.next
            # curnextnode
            cur.next = node
            # nodeprevcur
            node.prev = cur

    def search(self, item):
        """"""
        cur = self._head
        while cur != None:
            if cur.item == item:
                return True
            cur = cur.next
        return False
```

```python
class CycleArray:
    def __init__(self, size=1):
        self.size = size
        #  Python 
        self.arr = [None] * size
        # start 
        self.start = 0
        #  end 
        #  end 
        self.end = 0
        self.count = 0

    # 
    def resize(self, newSize):
        # 
        new_arr = [None] * newSize
        # 
        for i in range(self.count):
            new_arr[i] = self.arr[(self.start + i) % self.size]
        self.arr = new_arr
        #  start  end 
        self.start = 0
        self.end = self.count
        self.size = newSize

    #  O(1)
    def add_first(self, val):
        # 
        if self.is_full():
            self.resize(self.size * 2)
        #  start 
        self.start = (self.start - 1 + self.size) % self.size
        self.arr[self.start] = val
        self.count += 1

    #  O(1)
    def remove_first(self):
        if self.is_empty():
            raise Exception("Array is empty")
        #  start 
        self.arr[self.start] = None
        self.start = (self.start + 1) % self.size
        self.count -= 1
        # 
        if self.count > 0 and self.count == self.size // 4:
            self.resize(self.size // 2)

    #  O(1)
    def add_last(self, val):
        if self.is_full():
            self.resize(self.size * 2)
        #  end 
        self.arr[self.end] = val
        self.end = (self.end + 1) % self.size
        self.count += 1

    #  O(1)
    def remove_last(self):
        if self.is_empty():
            raise Exception("Array is empty")
        #  end 
        self.end = (self.end - 1 + self.size) % self.size
        self.arr[self.end] = None
        self.count -= 1
        # 
        if self.count > 0 and self.count == self.size // 4:
            self.resize(self.size // 2)

    #  O(1)
    def get_first(self):
        if self.is_empty():
            raise Exception("Array is empty")
        return self.arr[self.start]

    #  O(1)
    def get_last(self):
        if self.is_empty():
            raise Exception("Array is empty")
        # end  1
        return self.arr[(self.end - 1 + self.size) % self.size]

    def is_full(self):
        return self.count == self.size

    def size(self):
        return self.count

    def is_empty(self):
        return self.count == 0
```
